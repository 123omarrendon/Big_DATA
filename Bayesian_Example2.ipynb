# Fraud Detection\\ 
*using Synthetic Data (SMOTE)*

```python
# %pip install imbalanced-learn
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd
import numpy as np
```

```python
# Creating a mock imbalanced dataset
# Imagine 990 normal transactions and 10 fraudulent ones
data_dict = {
    'amount': np.random.uniform(10, 5000, 1000),
    'hour': np.random.randint(0, 24, 1000),
    'is_fraud': [1] * 10 + [0] * 990
}
df = pd.DataFrame(data_dict)
print(df)
X = df.drop('is_fraud', axis=1)
y = df['is_fraud']
```

```python
# Use SMOTE to generate synthetic fraud cases
# This analyzes the 'Small Data' (the 10 fraud cases) and creates new similar ones
smote = SMOTE(random_state=42, k_neighbors=3)
X_resampled, y_resampled = smote.fit_resample(X, y)

print(f"Original fraud count: {sum(y)}")
print(f"Resampled fraud count: {sum(y_resampled)}")
```

```python

# Train a classifier on the balanced data
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2)
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)

# Predict and Evaluate
predictions = clf.predict(X_test)
print(classification_report(y_test, predictions))
```

# Cyberattacks: 
*SMOTE for Zero-Day Detection*

```python
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix
import pandas as pd
import numpy as np
```

```python

# Mock Dataset: Features could be [packet_size, connection_duration, request_frequency]
# 5000 Normal logs (0), 50 Attack logs (1)
X = np.random.normal(size=(5050, 3)) 
y = np.array([0]*5000 + [1]*50)

# Apply SMOTE to increase 'Small Data' (Attacks) to match 'Normal'
# k_neighbors=5 looks at the 5 nearest real attacks to create a synthetic one
sm = SMOTE(random_state=42, k_neighbors=5)
X_res, y_res = sm.fit_resample(X, y)

print(f"Before SMOTE: {np.bincount(y)}")
print(f"After SMOTE: {np.bincount(y_res)}")
```

```python

# Now train a robust detector
model = GradientBoostingClassifier()
model.fit(X_res, y_res)

# Testing on a small set of real (not synthetic) data
test_data = np.random.normal(size=(10, 3))
print("Attack Probabilities:", model.predict_proba(test_data)[:, 1])
```
