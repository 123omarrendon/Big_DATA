```sql
-- Mock data
CREATE OR REPLACE TABLE workspace.bronze.customer_feedback AS SELECT * FROM VALUES
(1, "The new UI is amazing and very fast!"),
(2, "I hate the recent update, it keeps crashing.")
AS tab(id, review_text);


-- Analyze sentiment using built-in AI
SELECT 
  review_text, 
  ai_analyze_sentiment(review_text) AS sentiment
FROM workspace.bronze.customer_feedback;

-- https://www.databricks.com/product/pricing/ai-functions
```
